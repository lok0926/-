### 2. WHILE문

<p>● 반복 범위 없이 <span style='background-color: #fff5b1'>조건에 따라 반복 수행 여부를 결정하는 경우에 사용하는 반복문</span></p>
<p>● 조건문이 ‘참’인 동안에 while문 아래의 문장이 반복해서 수행됨.</p>



#### 1) 기본 구조
<p>● while 조건문:</p>
    <p>(들여쓰기) 반복 수행할 문장 입력</p>
    


#### 2) 특징
<p>● 조건문으로만 동작</p>
<p>● <span style='background-color: #fff5b1'>반복할 코드 안에 조건문에 영향을 주는 변화식이 들어감.</span></p>
<p>● 조건문이 ‘참’이면 반복할 코드와 변화식을 함께 수행함. ‘거짓‘이면 반복문을 끝내고 다음 코드를 실행함.</p>

```python
i=0 #i는 while문의 반복 루프에서 사용될 변수명, 0은 변숫값

while i<5: #while 조건문
    print(i) #수행할 문장(반복할 코드)
    i+=1 #변화식: i 값을 실행할 때마다 i의 값을 1씩 증가시키기 위해 추가한 코드
```

```python
i=0 #i는 while문의 반복 루프에서 사용될 변수명, 0은 변숫값

while i<5: #while 조건문
    print(i) #수행할 문장(반복할 코드)
    i=i+1 #i+=1과 동일함.
```

<span style="color:#0000ff"><strong>
    연습 1) while문을 이용하여 속담 '열 번 찍어 안 넘어가는 나무 없다.'가 아래와 같이 실현될 수 있도록 하시오.</strong>


```python
i = 1
while i <= 10:
    print(f"나무를 {i}번 찍었습니다.")
    if i == 10:
        print("나무 넘어가요~")
    i += 1
```

#### 3) 복합 연산자
<p> ● 다른 연산자와 할당 연산자(=)를 합쳐 놓은 연산자</p>



```python
i=100 #i는 while문의 반복 루프에서 사용될 변수명, 100은 변숫값

while i>0:
    print('Hello, world!', i)
    i-=1
```

### 3. WHILE문 제어하기

#### 1) 반복문 빠져나가기: break
<p>● while문은 조건문이 참인 동안 계속해서 내용을 반복하는데 <span style='background-color: #fff5b1'>반복을 강제로 멈추고 싶을 때 사용</span></p>



```python
i=0

while True:
    i+=1
    if(i>3):
        break #i의 값이 3을 넘어가면 반복을 멈추라는 의미
    print(i)
```

<p><span style="color:#0000ff"><strong>
    연습 4) while문을 활용해 300원짜리 커피를 최대 10잔만 팔 수 있는 자판기를 프로그래밍을 하시오.</strong></p>
<p><span style="color:red"><strong>※ 조건</span></strong></p>
<p> · 돈을 넣으면 '돈을 받았으니 커피를 줍니다.'라는 문장이 출력되어야 한다.</p>
<p> · 커피 1잔을 팔고 나면 '판매 가능한 커피는 ?잔입니다.'라는 문구가 출력되어야 한다.</p>
<p> · 커피 10잔을 다 팔고 나면 '커피가 다 떨어졌습니다. 판매를 중지합니다.'가 출력되어야 한다.</p>



```python
num = 10
while num > 0:
    n = input("커피는 300원인데 결재 하시겠습니까?\n")
    if n == "yes":
        print("돈을 받았으니 커피를 줍니다")
        num -= 1
        print(f"남은 커피 수는 {num}잔 입니다.")
    elif n == "no":
        print("안녕히가세용~")
print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
```

#### 2) 반복문 처음으로 돌아가기: continue
<p>● while문의 입력 조건과 맞지 않아 <span style='background-color: #fff5b1'>처음으로 돌아가고 싶을 때 사용</span></p>
<p>● 반복문 안에서 continue를 만나면 반복문으로 처음으로 돌아가 다음 반복을 진행함.</p>



```python
a=0

while a<10:
    a+=1
    
    if a%2==0:
        continue #a를 2로 나누었을 때 나머지가 0이면(a가 짝수인 경우) 맨 처음으로 돌아간다는 것을 의미
    print(a)
```

<p><span style="color:#0000ff"><strong>
    연습 5) 1부터 10까지의 숫자 중에서 3의 배수를 뺀 나머지 값을 출력하시오.

```python
a = 0
while a < 10:
    a += 1
    if a % 3 ==0:
        continue;
    print(a)
        
```
